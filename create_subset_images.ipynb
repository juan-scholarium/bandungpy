```python
import os
import fnmatch
import shutil
```

```python
root = "../../Datasets/stanford_online_products/images/"
pattern = "*.JPG"
im_dict = dict()

for path, subdirs, files in os.walk(root):
    for x in subdirs:
        im_dict[x] = list()
    for name in files:
        if fnmatch.fnmatch(name, pattern):
            get_dir = path.split("/")[-1]
            im_dict[get_dir].append(name)
```

```python
subset_image = "../../Datasets/stanford_online_products/subset_image/"

for k in list(im_dict.keys()):
    for i in range(1000):
        src = os.path.join(root, str(k + "/" + im_dict[k][i]))
        dst = os.path.join(subset_image, k)

        if os.path.isdir(os.path.join(subset_image, k)) == False:
            os.mkdir(os.path.join(subset_image, k))

        shutil.copy(src, dst)
```
