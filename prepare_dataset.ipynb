```python
import os
import fnmatch
import shutil
```

```python
import os

# buat folder dataset untuk menaruh dataset kita
dir_dataset = "./dataset"
if os.path.isdir(dir_dataset) == False:
    os.mkdir(dir_dataset)
```

```python
# extract dataset ke folder dataset
!unzip -nq Stanford_Online_Products.zip -d dataset/
print("Unzip Finish!")
```

```python
# cek isi folder dataset
!ls dataset/Stanford_Online_Products/
```

```python
# buat dictionary di mana key = folder_name, value = nama_file.jpg
root = "dataset/Stanford_Online_Products/"
pattern = "*.JPG"
im_dict = dict()

for path, subdirs, files in os.walk(root):
    for x in subdirs:
        im_dict[x] = list()
    for name in files:
        if fnmatch.fnmatch(name, pattern):
            get_dir = path.split("/")[-1]
            im_dict[get_dir].append(name)
```

```python
# kelompok gambar dan jumlah citra dari kelompok gambar tersebut
for k in list(im_dict.keys()):
    print(k, len(im_dict[k]))
```

```python
# ambill 1.000 citra dari setiap kelompok gambar
subset_dataset = "dataset/subset"

if os.path.isdir(subset_dataset) == False:
    os.mkdir(subset_dataset)

for k in list(im_dict.keys()):
    for i in range(1000):
        src = os.path.join(root, str(k + "/" + im_dict[k][i]))
        dst = os.path.join(subset_dataset, k)

        if os.path.isdir(os.path.join(subset_dataset, k)) == False:
            os.mkdir(os.path.join(subset_dataset, k))

        shutil.copy(src, dst)
```
