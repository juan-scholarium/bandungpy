## Berkenalan dengan filter (atau kernel) pada pengolahan citra digital

Mengapa perlu berkenalan dengan filter atau kernel pada pengolahan citra digital?

1. Filter atau kernel pada pengolahan citra digital menjadi dasar untuk memahami hubungan antara operasi konvolusi dengan ekstrasi fitur. Operasi konvolusi akan mengekstrak fitur dari citra masukan dengan memanfaatkan filter.
2. Jika kita memahami apa itu filter pada pengolahan citra digital, maka kita memiliki intuisi untuk memahami transfer learning.

Sebagai contoh, saya akan menggunakan dua jenis filter, yaitu filter Laplace dan filter Sobel

Filter Laplace untuk edge detection

![filter_laplace](https://wikimedia.org/api/rest_v1/media/math/render/svg/a58cb62fc6fb6a8ecf95edd1060c8a35d50963a4)

Sumber: https://en.wikipedia.org/wiki/Discrete_Laplace_operator

Filter Sobel untuk edge detection

![filter_sobel](https://wikimedia.org/api/rest_v1/media/math/render/svg/848abd56e0e33cf402f01183bfe1f68a93fb34a9)

Sumber: https://en.wikipedia.org/wiki/Sobel_operator

```python
# apply filter laplace dan sobel pada citra masukan
def apply_filters(im_input):
    im_laplace = filters.laplace(im_input)
    im_sobel_h = filters.sobel_h(im_input)
    im_sobel_v = filters.sobel_v(im_input)
    return im_laplace, im_sobel_h, im_sobel_v
```

### Ilustrasi operasi konvolusi antara filter dengan citra masukan

- Konvolusi adalah operasi yang menjumlahkan hasil dari perkalian matrix element wise antara filter / kernel dengan citra.
- Konvolusi akan menghasilkan output yang disebut dengan fitur atau convolved feature.
- Fitur ini digunakan oleh classifier (KNN, Random Forest, SVM, **Neural Network**) untuk mengklasifikasi citra.

Buat temen-temen yang tertarik belajar Machine Learning bisa cek link berikut ini: [Indonesia Belajar](https://youtu.be/Rk1VmHk_kUo)

### Hasil konvolusi antara filter dengan citra

```python
def plot_im(im1, im2):
    pyplot.figure(figsize=(15, 15))
    pyplot.subplot(121)
    pyplot.imshow(im1, cmap="seismic")
    pyplot.subplot(122)
    pyplot.imshow(im2, cmap="seismic")
    pyplot.show()
```

```python
im1_laplace, im1_sobel_h, im1_sobel_v = apply_filters(im1)
im2_laplace, im2_sobel_h, im2_sobel_v = apply_filters(im2)
```

Filter laplace

Citra keluaran dari filter laplace adalah garis tepi pada citra. Keluaran ini merupakan fitur (fitur garis tepi) yang diambil dari citra masukan.

```python
plot_im(im1_laplace, im2_laplace)
```

Filter sobel

Filter sobel mirip dengan filter laplace, yaitu mengambil fitur garis tepi dari citra masukan. Garis tepi terbagi menjadi dua, garis tepi horizontal dan garis tepi vertikal.

```python
plot_im(im1_sobel_h, im2_sobel_h)
```

```python
plot_im(im1_sobel_v, im2_sobel_v)
```

### Takeaway

Operasi konvolusi (antara filter dengan citra) akan mengekstrak fitur dari citra masukan. Pada contoh kali ini fitur yang diambil adalah garis tepi.

## Building block of Convolutional Neural Network (CNN)

CNN dapat dibagi menjadi dua bagian utama:
1. Feature learning / feature extraction : convolutional layer & pooling layer
2. Classification : fully connected layer / dense network / neural network

![cnn_flow](https://miro.medium.com/max/2510/1*vkQ0hXDaQv57sALXAJquxA.jpeg)

Sumber gambar: https://miro.medium.com/max/2510/1*vkQ0hXDaQv57sALXAJquxA.jpeg

Kesepakatan istilah:
- Shallower layer (lapisan lebih dangkal) adalah layer / lapisan yang dekat dengan input
- Deeper layer (lapisan lebih dalam) adalah layer / lapisan yang dekat dengan output

```python
import tensorflow as tf
print(tf.keras.__version__)
```

### Lapisan Convolution

Lapisan convolution tersusun atas banyak filter yang membentuk `3D volumes of neurons`. Bayangkan filter-filter seperti Laplace / Sobel namun dengan kombinasi angka-angka yang beragam dipakai untuk operasi konvolusi pada citra masukan. Masing-masing filter ini akan melakukan operasi konvolusi pada citra masukan dan memberikan keluaran fitur. Training model CNN akan mencari nilai dari setiap filter.

![convolutional_layer](https://cs231n.github.io/assets/cnn/depthcol.jpeg)

Sumber gambar: https://cs231n.github.io/convolutional-networks/

```python
# code pada tensorflow untuk convolution:
tf.keras.layers.Conv2D(filters=32,
                       kernel_size=(3, 3),
                       strides=(1, 1),
                       padding="same")
```

### Lapisan Aktivasi

Fungi lapisan aktivasi adalah memperkenalkan non-linearitas ke dalam neural network. Fungsi aktivasi yang umum digunakan adalah Rectifier Linear Unit (ReLU).

![](https://cs231n.github.io/assets/nn1/relu.jpeg)
Sumber gambar: https://cs231n.github.io/neural-networks-1/

```python
# code pada tensorflow untuk activation:
tf.keras.layers.Activation("relu")

# activation bisa ditempatkan bersamaan dengan Conv2D
tf.keras.layers.Conv2D(filters=32,
                       kernel_size=(3, 3),
                       strides=(2, 2),
                       padding="same",
                       activation="relu")
```

### Lapisan Pooling

Lapisan pooling akan mereduksi dimensi *convolved feature* sehingga beban komputasi berkurang. Pada contoh gambar dibawah ini operasi max pooling akan mengambil nilai maksimum dari wilayah operasi (2 x 2)

![pooling_layer](https://cs231n.github.io/assets/cnn/maxpool.jpeg)
Sumber gambar: https://cs231n.github.io/convolutional-networks/

```python
# code pada tensorflow untuk pooling:
tf.keras.layers.MaxPool2D(pool_size=(2, 2))
```

### Lapisan Fully Connected

Lapisan fully connected merupakan neural network yang bertugas mengklasifikasikan fitur yang sudah diekstrak ke dalam masing-masing kelas yang sudah diketahui sebelumnya. Untuk tugas klasifikasi, neuron pada lapisan output biasanya akan diikuti dengan activation `softmax`.

![fcl](https://miro.medium.com/max/700/1*kToStLowjokojIQ7pY2ynQ.jpeg)

Sumber gambar: https://miro.medium.com/max/700/1*kToStLowjokojIQ7pY2ynQ.jpeg

```python
# code pada tensorflow untuk dense / fully connected:
tf.keras.layers.Flatten()
tf.keras.layers.Dense(units=512, activation="relu")
tf.keras.layers.Dense(units=10, activation="softmax")
```

### Implementasi building block CNN dengan Keras

Terdapat dua cara untuk mendefinisikan model pada Keras
1. Menggunakan Sequential class
2. Menggunakan functional API

#### Mendefinisikan model dengan Sequential class

```python
# misalkan tugasnya adalaha klasifikasi bilangan 0 sampai 9

model = tf.keras.Sequential([

    # input
    tf.keras.Input(shape=(28, 28, 1)),

    # convolution & max pooling
    tf.keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation="relu"),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation="relu"),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),

    # fully connected layer
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=64, activation="relu"),
    tf.keras.layers.Dropout(rate=0.5),

    # output
    tf.keras.layers.Dense(10, activation="softmax"),
])

model.summary()
```

#### Mendefinisikan model dengan functional API

```python
# input
input_layer = tf.keras.Input(shape=(28, 28, 1))

# convolution & max pooling
x = tf.keras.layers.Conv2D(filters=32, kernel_size=(3, 3),
                           activation="relu")(input_layer)
x = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(x)
x = tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3),
                           activation="relu")(x)
x = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(x)

# fully connected layer
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(64, activation="relu")(x)
x = tf.keras.layers.Dropout(rate=0.5)(x)

# output
output_layer = tf.keras.layers.Dense(10, activation="softmax")(x)

model = tf.keras.Model(inputs=input_layer, outputs=output_layer)
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

### Takeaway

- Selama proses learning / training, CNN akan mempelajari filter-filter yang dibutuhkan untuk mengekstrak fitur dari citra masukan secara otomatis.
- Nilai matriks filter tidak perlu didefinisikan oleh seorang developer.
- Nilai pada filter dan neuron biasa dikenal dengan istilah bobot.

## Transfer Learning pada CNN

Melatih model CNN dapat dilakukan dengan dua pendekatan:
1. End to end learning: bobot model diinisialisasi secara acak.
2. Transfer learning: bobot model diinisialisasi dari model lain yang sudah dilatih sebelumnya.

Keuntungan menggunakan pendekatan transfer learning model lebih cepat konvergen dan memiliki akurasi lebih baik.

Sisi praktis ketika melatih model dengan pendekatan transfer learning:
1. Mula-mula kita mengganti bagian classifier pre-trained model (fully connected layer) dengan classifier baru.
2. Training classifier tersebut sambil membekukan bobot bagian convolution agar bobot bagian convolution tidak ikut berubah selama training bagian classifier.
3. Setelah training classifier selesai, kita bisa melakukan fine tuning terhadap bagian convolution

### The Dataset: Stanford online products dataset

https://cvgl.stanford.edu/projects/lifted_struct/

```python
dir_ds = "dataset/subset/train/"
pattern = "*.JPG"
im_dict = dict()

for path, subdirs, files in os.walk(dir_ds):
    for x in subdirs:
        im_dict[x] = list()
    for name in files:
        if fnmatch.fnmatch(name, pattern):
            get_dir = path.split("/")[-1]
            im_dict[get_dir].append(name)
```

```python
im_dict
```

```python
i = 1

pyplot.figure(figsize=(20,15))
for k in list(im_dict.keys()):
    fpath = dir_ds + k
    im = io.imread(fpath + "/" + im_dict[k][7])
    
    ax = pyplot.subplot(3, 4, i)
    pyplot.imshow(im)
    pyplot.title(k)
    pyplot.axis("off")
    
    i += 1
```
